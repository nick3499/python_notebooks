```python
#!/usr/bin/env python3
# Iron Age: Coins/Supplies/Goods Tally

'''
Enter technology title, e.g. Roof Tile Houses
Tally meant for current technology being unlocked.
If an amount entered falls below amount required, a negative int is returned.
Unlocked technologies are obsolete to this script.
Order of conditional statements based on top to bottom order in each column.
'''

technology = input('Enter Technology Title: ')

# Column 1
if technology == 'Roof Tile Houses':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining coins:', (int(coins_total) - 250))
    print('Remaining supplies:', (int(supplies_total) - 100))
else:
    pass

if technology == 'Architecture':
    stone_total = input('Enter your Stone total: ')
    print('Remaining Stone:', (int(stone_total) - 5))
else:
    pass

if technology == 'Militia':
    wine_total = input('Enter your Wine total: ')
    print('Remaining Wine:', (int(wine_total) - 8))
else:
    pass

# Column 2
if technology == 'Stock Breeding':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining Coins:', (int(coins_total) - 300))
    print('Remaining Supplies:', (int(supplies_total) - 250))
else:
    pass

if technology == 'Processions':
    dye_total = input('Enter your Dye total: ')
    print('Remaining Dye:', (int(dye_total) - 12))
else:
    pass

if technology == 'Processing':
    coins_total = input('Enter your Coins total: ')
    print('Remaining coins:', (int(coins_total) - 600))
else:
    pass

if technology == 'Iron Works':
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining Supplies:', (int(supplies_total) - 1500))
else:
    pass

if technology == 'Handicraft':
    coins_total = input('Enter your Coins total: ')
    print('Remaining coins:', (int(coins_total) - 1500))
else:
    pass

# Column 3
if technology == 'Agriculture':
    lumber_total = input('Enter your Lumber total: ')
    print('Remaining Lumber:', (int(lumber_total) - 30))
else:
    pass

if technology == 'Cottage':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining Coins:', (int(coins_total) - 600))
    print('Remaining Supplies:', (int(supplies_total) - 1250))
else:
    pass

if technology == 'Archery':
    marble_total = input('Enter your Marble total: ')
    print('Remaining Marble:', (int(marble_total) - 16))
else:
    pass

if technology == 'Sewerage':
    supplies_total = input('Enter your Supplies total: ')
    limestone_total = input('Enter your Limestone total: ')
    dye_total = input('Enter your Dye total: ')
    print('Remaining Supplies:', (int(supplies_total) - 750))
    print('Remaining Limestone:', (int(limestone_total) - 3))
    print('Remaining Dye:', (int(dye_total) - 15))
else:
    pass

# Column 4
if technology == 'Butchery':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining Coins:', (int(coins_total) - 100))
    print('Remaining Supplies:', (int(supplies_total) - 2500))
else:
    pass

if technology == 'Military Tactics':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining Coins:', (int(coins_total) - 1500))
    print('Remaining Supplies:', (int(supplies_total) - 2000))
else:
    pass

if technology == 'Fortification':
    supplies_total = input('Enter your Supplies total: ')
    ebony_total = input('Enter your Ebony total: ')
    stone_total = input('Enter your Stone total: ')
    print('Remaining Supplies:', (int(supplies_total) - 1400))
    print('Remaining Ebony:', (int(ebony_total) - 4))
    print('Remaining Stone:', (int(stone_total) - 20))
else:
    pass

# Column 5
if technology == 'Mathematics':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    print('Remaining Coins:', (int(coins_total) - 2000))
    print('Remaining Supplies:', (int(supplies_total) - 1500))
else:
    pass

if technology == 'Thermae':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    cloth_total = input('Enter your Cloth total: ')
    marble_total = input('Enter your Marble total: ')
    print('Remaining Coins:', (int(coins_total) - 600))
    print('Remaining Supplies:', (int(supplies_total) - 750))
    print('Remaining Cloth:', (int(cloth_total) - 11))
    print('Remaining Marble:', (int(marble_total) - 21))
else:
    pass

if technology == 'Plowing':
    lumber_total = input('Enter your Lumber total: ')
    iron_total = input('Enter your Iron total: ')
    print('Remaining Lumber:', (int(lumber_total) - 30))
    print('Remaining Iron:', (int(iron_total) - 13))
else:
    pass

if technology == 'Chain of Command':
    coins_total = input('Enter your Coins total: ')
    supplies_total = input('Enter your Supplies total: ')
    wine_total = input('Enter your Wine total: ')
    jewelry_total = input('Enter your Jewelry total: ')
    print('Remaining Coins:', (int(coins_total) - 600))
    print('Remaining Supplies:', (int(supplies_total) - 1100))
    print('Remaining Wine:', (int(wine_total) - 50))
    print('Remaining Jewelry:', (int(jewelry_total) - 16))
else:
    pass
```

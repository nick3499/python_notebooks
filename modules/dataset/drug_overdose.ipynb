# Accidental Drug Related Death Cases: 2012–June2017

## Import xml.etree.ElementTree

```python
import xml.etree.ElementTree as et
```

## Parse XML Dataset

Dataset URL: [https://data.ct.gov/api/views/rybz-nyjw/rows.xml](https://data.ct.gov/api/views/rybz-nyjw/rows.xml). **`wget`** can be used to download the XML data file: **`wget https://data.ct.gov/api/views/rybz-nyjw/rows.xml -O data.xml`**.

```python
pd = et.parse("data.xml")
```

## Find Root Element

```python
gr = pd.getroot()
```

## Find All Children Elements

```python
fr = gr.findall('row/row')
```

## Print Dataset Title and Total Records Value

```python
print u'\nAccidental Drug Related Death Cases: 2012\u2014June2017\n'
print 'total records:', len(fr), '\n'
```

## Iterate thru Rows of Data

**`if`** statements check missing elements

```python
for i in fr:
    if i.find('casenumber') is None:
        print 'case number      : N/A'
    else:
        print 'case number      :', i.find('casenumber').text
    if i.find('date') is None:
        print 'date             : N/A'
    else:
        print 'date             :', i.find('date').text[0:10]
    if i.find('sex') is None:
        print 'sex              : N/A'
    else:
        print 'sex              :', i.find('sex').text
    if i.find('race') is None:
        print 'race             : N/A'
    else:
        print 'race             :', i.find('race').text
    if i.find('age') is None:
        print 'age              : N/A'
    else:
        print 'age              :', i.find('age').text
    if i.find('residence_city') is None:
        print 'residence city   : N/A'
    else:
        print 'residence city   :', i.find('residence_city').text
    if i.find('residence_state') is None:
        print 'residence state  : N/A'
    else:
        print 'residence state  :', i.find('residence_state').text
    if i.find('residence_county') is None:
        print 'residence county : N/A'
    else:
        print 'residence county :', i.find('residence_county').text
    if i.find('death_city') is None:
        print 'death city       : N/A'
    else:
        print 'death city       :', i.find('death_city').text
    if i.find('death_county') is None:
        print 'death county     : N/A'
    else:
        print 'death county     :', i.find('death_county').text
    if i.find('location') is None:
        print 'location         : N/A'
    else:
        print 'location         :', i.find('location').text
    if i.find('injuryplace') is None:
        print 'injury location  : N/A'
    else:
        print 'injury location  :', i.find('injuryplace').text
    if i.find('immediatecausea') is None:
        print 'immediate cause  : N/A'
    else:
        print 'immediate cause  :', i.find('immediatecausea').text
    if i.find('mannerofdeath') is None:
        print 'manner of death  : N/A'
    else:
        print 'manner of death  :', i.find('mannerofdeath').text
    print u'\u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014'
```

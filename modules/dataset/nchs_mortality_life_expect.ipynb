# NCHS: Death Rates, Life Expectancy

## Import xml.etree.ElementTree

```python
import xml.etree.ElementTree as et
```

## Parse XML

```python
pd = et.parse("data.xml")
```

## Get Root Element

```python
gr = pd.getroot()
```

## Find Rows of Data

```python
fr = gr.findall('row/row')
```

## Print Title

```python
print '\nNCHS: Death Rates, Life Expectancy'
```

## Print Total Records

```python
print 'Total records:', len(fr), '\n'
```

## Iterate thru Rows of Data

The **`if`** statement handles the missing **`average_life_expectancy`** elements. A total of 8 were missing in the original data I used.

```python
for i in fr:
    print 'Year:', i.find('year').text
    print 'Race:', i.find('race').text
    print 'Sex:', i.find('sex').text
    if i.find('average_life_expectancy') is None:
        print 'Average Life Expectancy: N/A'
    else:
        print 'Average Life Expectancy:', i.find('average_life_expectancy').text
    print 'Mortality:', i.find('mortality').text
    print u'\u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014'
```

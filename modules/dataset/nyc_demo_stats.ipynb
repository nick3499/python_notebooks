# Python 2: lxml.etree Module: NYC Demographics Statistics: Iterate thru Zip Codes

## Import lxml.etree

```python
from lxml import etree as et
```

## Parse Local Data

```python
p = et.parse("data.xml")
```

## Find Statistics

```python
jn = p.findall("row/row/jurisdiction_name")
cp = p.findall("row/row/count_participants")
cf = p.findall("row/row/count_female")
pf = p.findall("row/row/percent_female")
cm = p.findall("row/row/count_male")
pm = p.findall("row/row/percent_male")
chl = p.findall("row/row/count_hispanic_latino")
phl = p.findall("row/row/percent_hispanic_latino")
canh = p.findall("row/row/count_asian_non_hispanic")
panh = p.findall("row/row/percent_asian_non_hispanic")
crpa = p.findall("row/row/count_receives_public_assistance")
prpa = p.findall("row/row/percent_receives_public_assistance")
cnpa = p.findall("row/row/count_nreceives_public_assistance")
pnpa = p.findall("row/row/percent_nreceives_public_assistance")
```

## Print Date

```python
print 'Data Courtesy of NYC OpenData'
print 'Created Jul 27, 2011'
print 'Updated Sep 22, 2017'
print u'\u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014'
```

## Iterate through Statistics

```python
for i in range(236):
    print 'Zip Code:', jn[i].text
    print 'Total Participants:', cp[i].text
    print 'Total Females:', cf[i].text
    print 'Percent Females:', pf[i].text
    print 'Total Males:', cm[i].text
    print 'Percent Males:', pm[i].text
    print 'Total Hispanic/Latino:', chl[i].text
    print 'Percent Hispanic/Latino:', phl[i].text
    print 'Total Asian:', canh[i].text
    print 'Percent Asian:', panh[i].text
    print 'Total on Public Assistance:', crpa[i].text
    print 'Percent on Public Assistance:', prpa[i].text
    print 'Total not on Public Assistance:', cnpa[i].text
    print 'Percent not on Public Assistance:', pnpa[i].text
    print u'\u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014'
```

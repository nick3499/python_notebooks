# Los Angeles: 2010 Census: by Zip Code

## Import lxml.etree

```python
from lxml import etree as et
```

## Parse Data

```python
p = et.parse("data.xml")
```

## Find Data Rows

```python
zc = p.findall("row/row/zip_code")
tp = p.findall("row/row/total_population")
ma = p.findall("row/row/median_age")
tm = p.findall("row/row/total_males")
tf = p.findall("row/row/total_females")
th = p.findall("row/row/total_households")
ahs = p.findall("row/row/average_household_size")
```

## Print Title

```python
print '\nLos Angeles: 2010 Census: by Zip Code\n'
```

## Print Rows

```python
for i in range(1, 319):
    print 'Zip Code:', zc[i].text
    print 'Total Population:', tp[i].text
    print 'Median Age:', ma[i].text
    print 'Total Males:', tm[i].text
    print 'Total Females:', tf[i].text
    print 'Total Households:', th[i].text
    print 'Average Household Size:', ahs[i].text
    print u'\u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014 \u2014'
```

# Iterate over DNA Sequences of Orchids in FASTA File Format

## Shebang/Interpreter Directive

```python
#!/usr/bin/env python2
```

## Import `Bio.SeqIO` Function from Biopython Module

Which means that it handles DNA sequence input/output.

```python
from Bio import SeqIO
```

## Iterate thru DNA Records

_Parse_ DNA sequences in a [FASTA](https://en.wikipedia.org/wiki/FASTA)-formatted file. See below.

`SeqIO.parse()`:
 - `ls_orchid.fasta` indicates _file handle_
 - `fasta` indicates _file format_
 - `i.description` returns sequence identifier, followed by descriptive string.
 - `repr(i.seq)` returns canonical string representation of indexed DNA sequence record

```python
for i in SeqIO.parse("ls_orchid.fasta", "fasta"):
    print "ID/Descrip.:", i.description
    print "Seq. Ref.  :", repr(i.seq)
```
